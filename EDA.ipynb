6. Use the IQR method to detect and remove outliers from the km_driven
column. Show before-and-after summary statistics.
(Include your Python code and output in the code box below.)

```python
import pandas as pd

## Load dataset
file_path = "/content/BIKE DETAILS.csv"
bike_df = pd.read_csv(file_path)

## --- Before removing outliers ---
print("Summary statistics of km_driven (Before removing outliers):\n")
print(bike_df['km_driven'].describe())

## IQR calculation
Q1 = bike_df['km_driven'].quantile(0.25)
Q3 = bike_df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

## Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

## Remove outliers
bike_df_no_outliers = bike_df[(bike_df['km_driven'] >= lower_bound) &
                              (bike_df['km_driven'] <= upper_bound)]

## After removing outliers
print("\nSummary statistics of km_driven (After removing outliers):\n")
print(bike_df_no_outliers['km_driven'].describe())
```

9. Generate a heatmap of the correlation matrix for all numeric columns.
What correlations stand out the most?
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## Load dataset
file_path = "/content/BIKE DETAILS.csv"
bike_df = pd.read_csv(file_path)

## One-hot encode seller_type to make it numeric
bike_df_encoded = pd.get_dummies(bike_df, columns=['seller_type'], drop_first=False)

## Select only numeric columns for correlation matrix
numeric_cols = bike_df_encoded.select_dtypes(include=['number'])

## Compute correlation matrix (only numeric columns)
corr_matrix = numeric_cols.corr()

## Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Correlation Matrix", fontsize=14)
plt.show()
```

10. Summarize your findings in a brief report:
 What are the most important factors affecting a bike's selling price?
 Mention any data cleaning or feature engineering you performed.
(Include your Python code and output in the code box below.)

```python
import pandas as pd

## Load dataset
file_path = "/content/BIKE DETAILS.csv"
bike_df = pd.read_csv(file_path)

## One-hot encode seller_type
bike_df_encoded = pd.get_dummies(bike_df, columns=['seller_type'], drop_first=False)

## Select only numeric columns for correlation matrix
numeric_cols = bike_df_encoded.select_dtypes(include=['number'])

## Compute correlation matrix
corr_matrix = numeric_cols.corr()

## Sort correlations with selling_price
price_corr = corr_matrix['selling_price'].sort_values(ascending=False)

## Report
print("===== Brief Report on Bike Selling Price Analysis =====\n")

print("1. Most Important Factors Affecting Selling Price:")
print(price_corr, "\n")

print("Key Findings:")
print("- Strongest positive correlation: 'ex_showroom_price' (higher showroom price  higher resale price).")
print("- Manufacturing 'year' is also positively correlated (newer bikes sell for more).")
print("- 'km_driven' has a weak negative correlation (more usage  slightly lower price).")
print("- 'seller_type' (Dealer/Individual) has minimal effect on price.\n")

print("2. Data Cleaning and Feature Engineering Performed:")
print("- Checked for missing values: 'ex_showroom_price' had ~41% missing values.")
print("- Outliers in 'km_driven' were removed using the IQR method for better reliability.")
print("- Converted categorical column 'seller_type' into numeric format using one-hot encoding.\n")

print("Overall Conclusion:")
print("""The most important factors influencing a bikes selling price are its original showroom price and year of manufacture, while usage (km driven)
also plays a smaller but notable role.Seller type does not significantly impact price.""")
```
